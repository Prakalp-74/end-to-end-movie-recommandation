```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ast
```

```python
df = pd.read_csv('movies_metadata.csv', low_memory=False)
```

```python
df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
movie = df[['id', 'title', 'overview', 'genres', 'tagline', 'popularity', 'vote_average', 'vote_count']]
```

```python
movie.isnull().sum()
```

```python
movie['overview'].head()
```

```python
movie = movie.dropna(subset=['title'])
```

```python
movie['overview'] = movie['overview'].fillna(' ')
```

```python
movie['tagline'] = movie['tagline'].fillna(' ')
```

```python
movie.duplicated().sum()
```

```python
movie = movie.drop_duplicates()
```

```python
movie.iloc[0]['genres']
```

```python
movie['genres'] = movie['genres'].apply(lambda x :" ".join([i['name'] for i in ast.literal_eval(x)]))
```

```python
movie.head()
```

```python
movie['tags'] = movie['overview'] + " " + movie['genres'] + " " + movie['tagline']
```
